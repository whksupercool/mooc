# 一、seaborn介绍

```python
iris = pd.read_csv("D:/project/python/mooc/数据处理/o25mso/homework/iris.csv")
iris.head()
```

```python
iris.Name.unique()
```

```python
color_map = dict(zip(iris.Name.unique(), ['blue', 'green', 'red']))
```

```python
for species, group in iris.groupby("Name"):
    plt.scatter(
        group['PetalLength'], 
        group['SepalLength'],
        color = color_map[species],
        alpha = 0.4,
        edgecolor=None,
        label=species
    )
plt.legend(frameon=True, title="Name")
plt.grid()
plt.xlabel("PetalLength")
plt.ylabel("SepalLength")
plt.show()
```

```python
sns.lmplot(
    x='PetalLength', 
    y='SepalLength', 
    data=iris, 
    hue="Name", 
    fit_reg=False
)
```

# 四、seaborn图形显示效果的设置

```python
x = np.linspace(0, 14, 100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.show()
    
sinplot()
```

```python
import seaborn as sns
sinplot()
```

### axes_style and set_style

```python
style = ["white", "dark", "whitegrid", "darkgrid", "ticks"]
sns.set_style(style[3])
sinplot()
```

```python
sns.axes_style()
```

```python
sns.set_style(style[3], {'grid.color': 'green'})
sinplot()
```

```python
sns.set()
sinplot()
```

# 五、seaborn调色功能

```python
def sin_plot():
    x = np.linspace(0, 14, 100)
    plt.figure(figsize=(10, 8))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label="sin(x+%s)*(%s+0.75)" % (i, i))
    plt.legend()
    plt.show()
```

```python
sin_plot()
```
