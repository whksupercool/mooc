```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

# 一、 matplotlib介绍

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

# 二、 matplotlib简单绘图之plot

```python
a = [1, 2, 3]
b = [4, 5, 6]
plt.plot(a, b)
plt.show()
```

```python
c = [10, 8, 5]
d = [1, 8, 3]
plt.plot(a, b, 'r--', c, d, 'b*')
```

```python
t = np.arange(0.0, 2.0, 0.01)
t.shape
```

```python
s = np.sin(t*np.pi)
s.shape
```

```python
plt.plot(t, s, 'r--', t*2, s)
```

```python
plt.plot(t, s, 'r--', label="aaa")
plt.plot(t*2, s, 'b', label="bbb")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("this is a demo")
plt.legend()
plt.show()
```

# 三、 matplotlib简单绘图之subplot

```python
plt.subplot(2, 1, 1)
plt.plot(t, s, 'r--')
plt.ylabel("y1")
plt.subplot(2, 1, 2)
plt.plot(t*2, s, 'b')
plt.ylabel("y2")
plt.xlabel("x")
plt.show()
```

```python
figure, ax = plt.subplots(2, 2)
ax[0][0].plot(t, s, 'r--')
ax[0][1].plot(t*2, s, 'b')
plt.show()
```

# 四、 pandas绘图之series

```python
s1 = Series([1, 2, 3, 4, 5])
```

```python
s1.cumsum()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s1.plot(kind="line", grid=True, label="s1", title="this is s1")
plt.legend()
plt.show()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
s1.plot(kind="line", grid=True, label="s1")
s2.plot(kind="line", grid=True, label="s2")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(s1)
ax[1].plot(s2)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
s1[0:10].plot(ax=ax[0], grid=True, kind="bar")
s2.plot(ax=ax[1], grid=True, label="s2")
plt.show()
```

# 五、pandas绘图之dataframe

```python
df = DataFrame(
    np.random.randint(1, 10, 40).reshape(10, -1),
    columns=["A","B","C","D"]
)
df
```

```python
df.plot(kind="bar", figsize=(10,8))
plt.show()
```

```python
df.plot(kind="bar", figsize=(10,8), stacked=True)
plt.show()
```

```python
df.plot(kind="area", figsize=(10,8), stacked=True)
plt.show()
```

```python
for i in df.index:
    df.iloc[i].plot(label=str(i))

plt.legend()
plt.show()
```

```python
df.plot()
plt.show()
```

```python
df.T.plot()
plt.show()
```
