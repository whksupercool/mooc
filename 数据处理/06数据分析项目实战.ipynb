# 二、股票市场分析实战--历史趋势分析

```python
alibaba = pdr.get_data_yahoo('BABA', start="2015-09-20", end="2020-12-31")
amazon = pdr.get_data_yahoo('AMZN', start="2015-09-20", end="2020-12-31")
```

```python
alibaba.head()
```

```python
amazon.head()
```

```python
alibaba["Adj Close"].plot(legend=True, figsize=(15, 8))
```

```python
amazon["Adj Close"].plot(legend=True, figsize=(15, 8))
```

```python
alibaba["Adj Close"].plot(legend=True, figsize=(15, 8))
amazon["Adj Close"].plot(legend=True, figsize=(15, 8))
```

```python
alibaba["high-low"] = alibaba["High"] - alibaba["Low"]
alibaba.head()
```

```python
alibaba["high-low"].plot(figsize=(15, 8))
```

```python
alibaba["daily-return"] = alibaba["Adj Close"].pct_change()
```

```python
alibaba.head()
```

```python
alibaba["daily-return"].plot(figsize=(20, 8), linestyle='--', marker='o')
```

```python
alibaba["daily-return"].plot(figsize=(10, 8), kind='hist')
```

```python
sns.displot(alibaba["daily-return"].dropna(), kde=True, bins=50, color='green')
```

# 三、股票市场分析实战--风险分析

```python
company = ["AAPL","AMZN","FB","GOOG","MSFT"]
top_tech_df = pdr.get_data_yahoo(company, start="2015-09-20", end="2020-12-31")["Adj Close"]
```

```python
top_tech_df.head()
```

```python
top_tech_dr = top_tech_df.pct_change()
```

```python
top_tech_dr.head()
```

```python
top_tech_df.plot()
```

```python
top_tech_df[["AAPL","FB","MSFT"]].plot()
```

```python
sns.jointplot(x="GOOG", y="GOOG", data=top_tech_dr, kind="scatter")
```

```python
sns.jointplot(x="AMZN", y="GOOG", data=top_tech_dr, kind="scatter")
```

```python
sns.jointplot(x="MSFT", y="FB", data=top_tech_dr, kind="scatter")
```

```python
sns.pairplot(top_tech_dr.dropna())
```
