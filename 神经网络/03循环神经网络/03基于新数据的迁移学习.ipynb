```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from keras.models import Sequential
from keras.layers import Dense
from joblib import dump, load
```

```python
X = np.linspace(-5, 5, 101)
y = X ** 2

X2 = np.linspace(-5, 5, 101)
y2 = X2 ** 2 + 10

fig = plt.figure(figsize=(7, 5))
plt.scatter(X, y, label='X_y')
plt.scatter(X2, y2, label="X2_y2")
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
X = np.array(X).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
print(X.shape)
print(y.shape)
```

```python
model_1 = Sequential()
model_1.add(Dense(units=50, input_dim=1, activation='relu'))
model_1.add(Dense(units=50,activation='relu'))
model_1.add(Dense(units=1,activation='linear'))
model_1.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model_1.summary()
```

```python
%%time
model_1.fit(X, y, epochs=100)
```

```python
y_predict = model_1.predict(X)
```

```python
fig2 = plt.figure(figsize=(7, 5))
plt.scatter(X, y)
plt.plot(X, y_predict, 'r')
plt.title('y vs x(epoch=100)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig3 = plt.figure(figsize=(7, 5))
plt.scatter(X, y)
plt.plot(X, y_predict, 'r')
plt.title('y vs x(epoch=200)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
